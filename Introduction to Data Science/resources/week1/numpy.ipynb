```python
# Sequence of floats
# 2nd argument is inclusive
# 3rd argument is the number of numbers you want to generate
h = np.linspace(0, 2, 15)
print(h)
```

# Array Operations

```python
a = np.array([10, 20, 30, 40])
b = np.array([1, 2, 3, 4])
c = a + b
d = a - b
e = a * b
f = a / b
print(c)
print(d)
print(e)
print(f)
# Must be the same number elements
```

```python
# Converting temperatures
farenheit = np.array([0, -10, -5, -15, 0])
celcius = (farenheit - 31) * (5/9)
print(celcius)
```

```python
# Boolean array - meeting conditions
bool = celcius > -20
print(bool)
```

```python
even = celcius % 2 == 0
print(even)
```

```python
# Matrix manipulation
A = np.array([[1, 1], [0, 1]])
B = np.array([[2, 0], [3, 4]])
# elementwise product
# each position with its
# corresponding position
C = A * B
print(C)
# matrix product - dot product
# Sum(A row * B col) -> n
D = A @ B
print(D)
```

```python
# Linear Algebra
''' The product of two matrices is only plausible when the inner dimensions of the two matrices is the same. Dimensions: number of elements both horizontally and vertically in the  matrices
'''
A.shape
```

```python
# Upcasting
''' 
the types of the arrays are converted to the more general of the two types
'''
array1 = np.array([1, 2, 3])
print(array1.dtype)
array2 = np.array([1.1, 2.2, 3.3])
print(array2.dtype)
array3  = array1 @ array2
print(array3.dtype)
```

```python
# Agregation functions
print(array1.sum())
print(array1.max())
print(array3.min())
print(array2.mean())
```

```python

# Display images
from PIL import Image
from IPython.display import display

im = Image.open('gabo.jpg')
display(im)

```

```python
imgArray = np.array(im)
print(imgArray.shape)
print(imgArray.dtype)
#uint8 8 bits 256 - no negatives
```

```python
#Invert color image
mask = np.full(imgArray.shape, 255)
modifiedArray = imgArray - mask
modifiedArray = modifiedArray*-1
modifiedArray = modifiedArray.astype(np.uint8)
invImg = Image.fromarray(modifiedArray)
display(invImg)
```

# Indexing Slicing Iterating

## Indexing

```python
a = np.array([1, 2, 3])
a[2]
```

```python
a = np.array([[1, 2], [3, 4], [5, 6]])
a[1][1]
```

```python
b = np.array([a[0][0], a[1][1], a[2,1]])
b
```

```python
c = (a[[0, 1, 2], [0, 1, 1]])
c
```

## Boolean Indexing

```python
print(a > 5)
```

```python
print(a[a > 5])
```

## Slicing

```python
a = np.array([0, 1, 2, 3,  4, 5])
print(a[:3])
print(a[2:4])
```

```python
a = np.array([[0, 1, 2, 4], [5, 6, 7, 8], [9, 10, 11, 12]])
a[:2]
```

```python
a[:2, 1:3]
```

```python
# Arrays are modified by reference and not copy
sub_array = a[:2, 1:3]
print(sub_array[0,0])
sub_array[0,0] = 50
print(sub_array[0,0])
print(a[0,1])
```
